```python
a = nba.loc[nba['Salary'] > 5000000, ['Name', 'Salary']]
a['Salary'].max()
```

```python
a['Salary'].iloc[:10]
```

```python
nba.loc[(nba['Salary'] < 2000000) & (nba['Age'] < 25), ['Name', 'Salary']]
```

```python
a.to_csv('nba_young.csv') # writing to file
```
