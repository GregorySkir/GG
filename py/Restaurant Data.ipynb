## Joining different columns
The INNER JOIN keyword selects records that have matching values in both tables, based on a key

```python
week1.merge(week2, how="inner", on="Customer ID") #the parameter on works only if the column in both df are identical
# FOOD ID_x represent week1 and the y represent the week2
```

```python
week1.merge(week2, how="inner", on="Customer ID", suffixes=("_week 1", "_week 2")).head()
# customer 155 appears twice, with the same food_id of week2
```

```python
week1[week1["Customer ID"] == 155]
```

```python
week2[week2["Customer ID"] == 155]
```

```python
week1.merge(week2, how="inner", on=["Customer ID","Food ID"]) 
# both conditions in "on" parameter must be fulfill
# customer id num 21 appears twice but its not a duplicate its one for every week as you can see
```

```python
week1[week1["Customer ID"] ==21]
```

```python
week2[week2["Customer ID"] ==21]
```

###  Left JOIN
use only keys from left frame, similar to a SQL left outer join; preserve key order.
That is, all of the rows of week1 are in the new dataframe

```python
week1.merge(week2, how="left", on = "Customer ID", suffixes=["_week1", "_week2"])#.apply(len)
# if you apply(len) you can see that the new dataframe contains only 454 row which means that we have valid values in both
# week1 and week2 which mean that there is 46 customers that cames in two weeks
```

```python
left = week1.merge(food, how = "left", on = "Food ID")
left
```

```python
week2.merge(customer_id, how= "left", left_on= "Customer ID", right_on= "ID").head()
```

```python
week2.merge(customer_id, how= "left", left_on= "Customer ID", right_on= "ID").drop("ID", axis =1).head()
```
