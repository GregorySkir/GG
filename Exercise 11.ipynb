1. Load the data from the file fortune1000.csv that is attched to this assignment. 

```python
fort = pd.read_csv("fortune1000.csv")
```

```python
fort.shape
```

2. b. Print the first 5 rows of the dataframe

```python
fort.head()
```

3. What is the average revenue of all of the companies? What is the average profit?

```python
fort['Revenue'].mean()
```

4. Who are the 7 most successful companies (by revenue)?

```python
fort.sort_values("Revenue", ascending=False).head(7)
```

5. Who are the companies with more than 300,000 employees?

```python
fort.loc[fort['Employees'] > 300000]
```

6. How many sectors are in the database? How many industries?

```python
sector=len(pd.unique(fort['Sector']))
print(sector)
industry=len(pd.unique(fort['Industry']))
print(industry)
```

7. a. What are the 5 most popular industries (by companies)? How many companies are associated with these industries?

```python
f=fort.groupby(['Industry'])['Company'].count().sort_values(ascending=False)
f.head()
```

7.b. For each industry, which is the most profitable company? (you might want to use idxmax).

```python
fort.loc[(fort.groupby(['Industry'])["Profits"].idxmax()),['Company','Industry', 'Profits']]
```

8. a. Split the location column into two new columns named 'City' and 'State'. 
<br>- Use str functions. 
<br>- You can also use unpacking into these two columns

```python
fort['City'], fort['State'] = fort['Location'].str.split(', ', 1).str
fort.head()
```

8. b. Does every state in the USA (out of 50) has a company in the database? Which state is the most popular one?

```python
print('Number of states ' + str(len(pd.unique(fort['State']))))
fort.State.value_counts()[:5]
```

9. What are the names of the companies which are located in NY and have a rank lower than 50?

```python
fort.loc[(fort['Rank'] < 50) & (fort['State'] == 'NY'), ['Company', 'Rank']]
```

10. Which sector have (on average) losses? Which sector is the most profitable?

```python
fort.groupby(['Sector'])['Profits'].agg(['max', 'min'])
```

11. a. Add a new column to the database after the Profits column called Rev/Emp which holds the revenues each employee brings to the company.

```python
fort['Rev/Emp'] = fort['Profits']/fort['Employees']
fort['Rev/Emp']
```

11. b. For each sector and state, what is the count, mean and std of Rev/Emp?

```python
fort.groupby(['Sector', "State"])['Rev/Emp'].agg(['count', 'mean', 'std'])
```

12. We want to add for each company its Ticker. To do that, join the data from the file "compaines.csv" with your original data, and only keep the column 'ticker' from the new data. The data should be joined based on the columns "Company", in the fotune1000 database and "short name", in the companies data base. 
<br> Make sure to change the column name to "Ticker".


```python
comp = pd.read_csv("companies.csv")
comp.rename(columns={"short name": "Company", "ticker": "Ticker"}, inplace=True)
res = fort.merge(comp[["Company", "Ticker"]], how="left", on= "Company")
res
```

13. Save your updated database to a new file and name it fortune1000_extended.csv.

```python
res.to_csv('fortune1000_extended.csv')
```
